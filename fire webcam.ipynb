```python
import cv2
import numpy as np
import tensorflow as tf
import pyttsx3
from tensorflow.keras.preprocessing import image

# ✅ Load the trained fire detection model
model = tf.keras.models.load_model("fire_detection.keras")

# ✅ Initialize text-to-speech engine
engine = pyttsx3.init()
engine.setProperty("rate", 150)  # Set speaking speed

# ✅ Function to preprocess webcam frames
def preprocess_frame(frame):
    img = cv2.resize(frame, (224, 224))  # Resize to model input size
    img = img.astype("float32") / 255.0  # Normalize
    img = np.expand_dims(img, axis=0)  # Convert to batch format
    return img

# ✅ Start webcam capture
cap = cv2.VideoCapture(0)  # 0 for default webcam

# ✅ Flag to avoid repetitive voice alerts
fire_alert_triggered = False

while True:
    ret, frame = cap.read()
    if not ret:
        print("⚠️ Failed to grab frame. Exiting...")
        break

    # 🔹 Preprocess the frame
    processed_frame = preprocess_frame(frame)

    # 🔹 Get prediction
    prediction = model.predict(processed_frame)[0][0]  # Extract confidence score

    # 🔥 Fire Detected
    if prediction > 0.5:
        label = f"🔥 Fire Detected ({prediction:.2f})"
        color = (0, 0, 255)  # Red

        # 🎤 Voice Alert (Only once per detection)
        if not fire_alert_triggered:
            engine.say("Warning! Fire detected!")
            engine.runAndWait()
            fire_alert_triggered = True  # Avoid multiple alerts

    else:
        label = f"✅ No Fire ({prediction:.2f})"
        color = (0, 255, 0)  # Green
        fire_alert_triggered = False  # Reset when fire disappears

    # 🔹 Draw bounding box & label
    cv2.rectangle(frame, (50, 50), (500, 100), color, -1)  # Background rectangle
    cv2.putText(frame, label, (60, 90), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)  # Text label

    # 🔹 Display the frame
    cv2.imshow("🔥 Fire Detection - Webcam", frame)

    # Press 'q' to quit and properly release webcam
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# ✅ Release resources
cap.release()
cv2.destroyAllWindows()
```
